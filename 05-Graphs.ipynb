```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mu, sigma = 100, 15
dataset = mu+sigma*np.random.randn(1000)

#this histogram of the data
n, bins, patches = plt.hist(dataset, 50, normed='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, 0.025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

# Random Low and High Temperature Data

```python
data_set_size = 15

low_mu, low_sigma = 50, 4.3
low_data_set = low_mu + low_sigma * np.random.randn(data_set_size)

high_mu, high_sigma = 57, 5.2
high_data_set = high_mu + high_sigma * np.random.randn(data_set_size)

days = list(range(1, data_set_size + 1))
```

```python
plt.plot(days, low_data_set)
```

```python
plt.plot(days, low_data_set, 
         days, high_data_set)
```

```python
plt.plot(days, low_data_set, 
         days, low_data_set, "vm",
         days, high_data_set,
         days, high_data_set, "^k")
```

```python
plt.plot(days, low_data_set,
        days, high_data_set)

plt.xlabel('Day')
plt.ylabel('Temperature: degrees Farenheit')
plt.title('Randomized temperature data')
```

```python
plt.plot(
        days, high_data_set)

plt.xlabel('Day')
plt.ylabel('Temperature: degrees Farenheit')
plt.title('Randomized temperature data')
```

```python
t1 = np.arange(0.0, 2.0, 0.1)
t2 = np.arange(0.0, 2.0, 0.01)

# note that plot returns a list of lines.  The "l1, = plot" usage
# extracts the first element of the list into l1 using tuple
# unpacking.  So l1 is a Line2D instance, not a sequence of lines
l1, = plt.plot(t2, np.exp(-t2))
l2, l3 = plt.plot(t2, np.sin(2 * np.pi * t2), '--o', t1, np.log(1 + t1), '.')
l4, = plt.plot(t2, np.exp(-t2) * np.sin(2 * np.pi * t2), 's-.')

plt.legend((l2, l4), ('oscillatory', 'damped'), loc='upper right', shadow=True)
plt.xlabel('time')
plt.ylabel('volts')
plt.title('Damped oscillation')
plt.show()
```

# Tick Marks, Labels, and Grids

```python
points = 1000

fig = plt.figure()

sub_plot1 = fig.add_subplot(1, 1, 1)
sub_plot1.plot(np.random.rand(points).cumsum())

number_of_ticks = 5
ticks = np.arange(0, points, points//number_of_ticks)
sub_plot1.set_xticks(ticks)

labels = sub_plot1.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation = 45, fontsize = 'small')

sub_plot1.set_title("My First Ticked Plot")
sub_plot1.set_xlabel("Groups")

sub_plot1.grid(True)
gridlines = sub_plot1.get_xgridlines() + sub_plot1.get_ygridlines()
for line in gridlines:
    line.set_linestyle(':')
plt.show()
```

# Plot Annotations

```python
data_set_size = 10

fig = plt.figure()
sub_plot1 = fig.add_subplot(1, 1, 1)
sub_plot1.plot(np.random.rand(data_set_size).cumsum())

sub_plot1.text(1, 0.5, r'an equation: $E=mc^2$', fontsize = 18, color = 'red')
sub_plot1.text(1, 1.5, "Hello, Mountain Climbing!", family = 'monospace', fontsize = 14, color = 'green')

sub_plot1.text(0.5, 0.5, "We are centered, now", transform = sub_plot1.transAxes)

sub_plot1.annotate('shoot arrow', xy=(2,1), xytext=(3,4), 
                   arrowprops=dict(facecolor='red', shrink=0.05))
plt.show()
```

```python
x = np.arange(0, 10, 0.005)
y = np.exp(-x/2.) * np.sin(2*np.pi*x)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)
plt.show()
```
