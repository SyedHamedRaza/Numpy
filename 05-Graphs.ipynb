# Graphs

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mu, sigma = 100, 15
dataset = mu+sigma*np.random.randn(1000)

#this histogram of the data
n, bins, patches = plt.hist(dataset, 50, normed='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, 0.025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
